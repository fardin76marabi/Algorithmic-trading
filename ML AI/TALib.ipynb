```python
import yfinance as yf
from datetime import datetime, timedelta
import time

# تنظیمات yfinance
yf.set_config('proxy', None)  # غیرفعال کردن پراکسی

# بررسی نسخه yfinance
print(f"نسخه yfinance: {yf.__version__}")

# تعریف نماد
ticker = "AAPL"

# تنظیم بازه زمانی (90 روز پیش تا دیروز)
end_date = datetime.now() - timedelta(days=1)  # تا دیروز
start_date = end_date - timedelta(days=90)    # 90 روز پیش

start_date_str = start_date.strftime('%Y-%m-%d')
end_date_str = end_date.strftime('%Y-%m-%d')
print(f"دریافت داده‌ها از {start_date_str} تا {end_date_str}")

# تلاش برای دریافت داده‌ها
try:
    apple_data = yf.download(ticker, start=start_date_str, end=end_date_str, auto_adjust=False, progress=False)
    
    if not apple_data.empty:
        print("داده‌های دریافت‌شده:")
        print(apple_data.head())
    else:
        print("هیچ داده‌ای دریافت نشد. بازه زمانی یا اتصال را بررسی کنید.")
except Exception as e:
    if "429" in str(e):
        print("خطای 429: تعداد درخواست‌ها زیاد است. 10 ثانیه صبر کنید...")
        time.sleep(10)
        apple_data = yf.download(ticker, start=start_date_str, end=end_date_str, auto_adjust=False, progress=False)
        if not apple_data.empty:
            print("داده‌های دریافت‌شده پس از تلاش مجدد:")
            print(apple_data.head())
        else:
            print("هیچ داده‌ای دریافت نشد.")
    else:
        print(f"خطا رخ داد: {e}")
```

```python
import talib
import numpy as np

high = np.array(data['High'], dtype=float)
low = np.array(data['Low'], dtype=float)

high = high.flatten()
low = low.flatten()


talib.AROONOSC(high,low,timeperiod=14)
```

```python
open = np.array(data['Open'], dtype=float)
close = np.array(data['Close'], dtype=float)
high = np.array(data['High'], dtype=float)
low = np.array(data['Low'], dtype=float)

high = high.flatten()
low = low.flatten()
close=close.flatten()
open=open.flatten()

talib.CDLENGULFING(open,high,low,close)
```
